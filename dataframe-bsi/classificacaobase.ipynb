```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
import shap
import joblib

# Carregar os dados
#df_dados = pd.read_csv('df_ORIGINAL_328.csv', sep=',')
#df_dados = pd.read_csv('df_ORIGINAL_148.csv', sep=',')
#df_dados = pd.read_csv('df_CONJUNTO_A.csv', sep=',')
#df_dados = pd.read_csv('df_CONJUNTO_B.csv', sep=',')
df_dados = pd.read_csv('df_CONJUNTO_C.csv', sep=',')

# Substituir vírgulas por pontos em todas as colunas que podem conter números representados como strings
df = df.replace(',', '.', regex=True)

# Separar a variável alvo (status) e os preditores
X = df.drop(columns=['status'])
y = df['status']

# Padronizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=95, stratify=y)

# Configurar a validação cruzada estratificada com 10 pastas
kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=95)

# Definir os classificadores
classifiers = {
    'Logistic Regression': LogisticRegression(random_state=95, max_iter=2000, solver='liblinear', C=1.0),
    'SVM': SVC(probability=True, random_state=95, kernel='rbf'),
    'Random Forest': RandomForestClassifier(random_state=95, n_estimators=100),
    'AdaBoost': AdaBoostClassifier(random_state=95, n_estimators=100),
    'Neural Network': MLPClassifier(random_state=95, max_iter=2000),
    'QDA': QuadraticDiscriminantAnalysis(),
    'Gaussian Process': GaussianProcessClassifier(kernel=RBF(length_scale=1.0), random_state=95),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Decision Tree': DecisionTreeClassifier(random_state=95),
    'Naive Bayes': GaussianNB()
}

# Avaliar cada classificador
results = {}
base_results = {'sua_base': 'df_CONJUNTO_C', 'base_pca': 'PCA aplicada'}

for name, clf in classifiers.items():
    print(f"\nModelo: {name}")
    
    # Validação cruzada
    cv_results = cross_validate(
        clf,
        X_train,
        y_train,
        cv=kf,
        scoring=['accuracy', 'precision_weighted', 'recall_weighted', 'f1_weighted'],
        return_train_score=False
    )
    
    accuracies = [float(round(acc, 4)) for acc in cv_results['test_accuracy']]
    print(f"Acurácias formatadas: {', '.join(map(str, accuracies))}")
    print(f"Precision média: {cv_results['test_precision_weighted'].mean():.4f}")
    print(f"Recall média: {cv_results['test_recall_weighted'].mean():.4f}")
    print(f"F1-score médio: {cv_results['test_f1_weighted'].mean():.4f}")

    # Treinar o modelo
    clf.fit(X_train, y_train)

    # Fazer previsões no conjunto de teste
    y_pred = clf.predict(X_test)
    
    # Avaliar o modelo no conjunto de teste
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    
    print(f"\nAcurácia no conjunto de teste: {accuracy:.4f}")
    print("Relatório de Classificação no conjunto de teste:")
    print(report)

    # Exibir a matriz de confusão
    conf_matrix = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=clf.classes_)
    disp.plot(cmap=plt.cm.PuRd, colorbar=False)
    plt.title(f'{name} - Matriz de Confusão')
    plt.show()

    # Armazenar os resultados
    results[name] = {
        'accuracy': accuracy,
        'classification_report': report,
        'confusion_matrix': conf_matrix,
        'cross_val_accuracies': accuracies
    }

    # Salvar o modelo
    joblib.dump(clf, f"{name.replace(' ', '_').lower()}_model.pkl")
    base_results[f"resul_{name.split(' ')[0].lower()}"] = accuracies

# Salvar os resultados em um arquivo CSV
base_results_df = pd.DataFrame([base_results])
base_results_df.to_csv('base_results.csv', mode='a', index=False, header=not pd.io.common.file_exists('base_results.csv'))

# Análise com SHAP para Logistic Regression como exemplo
explainer = shap.LinearExplainer(classifiers['Logistic Regression'], X_train)
shap_values = explainer.shap_values(X_test)

shap.summary_plot(shap_values, X_test, feature_names=X.columns)
shap.summary_plot(shap_values, X_test, feature_names=X.columns, plot_type="bar")
```
