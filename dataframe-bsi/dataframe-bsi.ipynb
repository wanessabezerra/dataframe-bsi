# DataFrame BSI

Esse dataframe é a junção dos dataframes turma, matrícula, componentes curriculares  e dados pessoais dos discentes de Sistemas de Informação da UFRN, no período de 2009.1 à 2022.2.

Importando o Pandas e o Csv.

```python
import pandas as pd
import csv
```

Leitura dos arquivos em csv e carregamento desses dados em um DataFrame usando o ponto e vírgula como separador.

```python
dfturma = pd.read_csv('turmas-2009-2022.csv', sep=';')
dfcomponente = pd.read_csv('componentes-curriculares-presenciais.csv', sep=';')
dfmatricula = pd.read_csv('matricula-2009-2022.csv', sep=';')
dfpessoal = pd.read_csv('dados-pessoais.csv', sep=';')
```

## Junção dos Dataframes: Turma, Matrícula, Componentes Curriculares e Dados Pessoais

Junção do dataframe **matrícula** com o dataframe **turma** usando como chave *id_turma.*

```python
turma_matricula = pd.merge(dfturma, dfmatricula, on='id_turma', how='inner')
turma_matricula
```

Junção do dataframe **turma_matricula** com o dataframe **componente** usando as chaves *id_componente_curricular* e *id_componente*.

```python
turma_matricula_componente = pd.merge(turma_matricula, dfcomponente, left_on="id_componente_curricular", right_on="id_componente", how='inner')
turma_matricula_componente 
```

```python
turma_matricula_componente_pessoal22 = turma_matricula_componente['discente'].unique()
turma_matricula_componente_pessoal22
```

Junção do dataframe **turma_matricula_componente** com o dataframe **pessoal** usando as chaves *discente* e *id_discente*.

```python
turma_matricula_componente_pessoal = pd.merge(turma_matricula_componente, dfpessoal , left_on="discente", right_on="id_discente", how='inner')
turma_matricula_componente_pessoal
```

```python
turma_matricula_componente_pessoal = turma_matricula_componente_pessoal['discente'].unique()
turma_matricula_componente_pessoal
```

## Selecionando Colunas

Listando colunas.

```python
turma_matricula_componente_pessoal.columns
```

```python
turma_matricula_componente_pessoal = turma_matricula_componente_pessoal.rename(columns={'numero_total_faltas': 'frequencia'})
```

Reduzindo as colunas.

```python
turma_matricula_componente_pessoal = turma_matricula_componente_pessoal.filter(items=['discente', 'unidade', 'nota', 'media_final', 'frequencia', 'descricao', 'ano', 'id_componente', 'nome', 'ch_total', 'sexo', 'ano_nascimento', 'ano_ingresso', 'ano_ingresso', 'status'])
turma_matricula_componente_pessoal
```

## Limpeza de dados

Removendo linhas duplicadas.

```python
turma_matricula_componente_pessoal = turma_matricula_componente_pessoal.drop_duplicates(subset=['discente', 'unidade', 'media_final', 'descricao', 'ano', 'id_componente', 'nome'])
```
